```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('data.csv',encoding='cp949')
df['날짜'] = pd.to_datetime(df['날짜'])
df = df.fillna(0)
df
```

```python
timeSeries=df
timeSeries.index=timeSeries.날짜
ts=timeSeries.drop("날짜", axis=1)
ts
```

```python
ts.shape
```

```python
ts[['PM10']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
ts[['PM2.5']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
ts[['평균기온']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
ts[['강수량']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
ts[['풍속']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
ts[['풍향']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
ts[['상대습도']].plot(kind = 'line', figsize = (15,6), linewidth = 3, fontsize = 20)
plt.xlabel('Date', fontsize=13)
plt.ylabel('', fontsize=13)
plt.show()
```
