```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv',encoding='cp949')
df['날짜'] = pd.to_datetime(df['날짜'])
df = df.fillna(0)
df.rename(columns={'날짜':'Date','오존':'Ozone', '이산화질소':'NO2','일산화탄소':'CO','아황산가스':'SO2','평균기온':'Temp','강수량':'Rain','풍속':'Wind Speed','풍향':'Wind Direction','상대습도':'Humid'},inplace=True)
df.drop(['PM2.5'],axis=1, inplace=True)
df
```

데이터셋 분리하기(training and test)

```python
from sklearn.model_selection import train_test_split
x=df[['Ozone','NO2','CO','SO2','Temp','Rain','Wind Speed','Wind Direction','Humid']]
y=df[['PM10']]

x_train, x_test, y_train,y_test=train_test_split(x,y,train_size=0.8,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
my_air=[[0.2,0.03,0.2,0.005,17,0,2.2,50,53]]
my_predict=mlr.predict(my_air)
```

```python
y_predict=mlr.predict(x_test)
```

```python
plt.scatter(y_test,y_predict,alpha=0.4)
plt.xlabel("Actual Air")
plt.ylabel("Predicted Air")
plt.title("Multiple Linear Regression")
plt.show()
```

```python
print(mlr.coef_)
```
